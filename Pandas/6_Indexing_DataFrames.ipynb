```python
import pandas as pd
import numpy as np
```

```python
# Loading the DataFrame
df = pd.read_csv("Admission_Predict.csv", index_col = 0)

# Extracting column names, striping white space and then turning then lowercase
df.columns = [x.lower().strip() for x in df.columns]

df.head(3)
```

```python
# Changing Index

# The goal is to change the index column from "Serial No."" to "chance of admit". But, while doing so
# the "Serial No." column data should not be lost.

df['Serial No.'] = df.index
df = df.set_index('chance of admit')
df.head(3)
```

```python
# Removing Index

df = df.reset_index()
df.head(3)
```

### Multi-Level Index

```python
df = pd.read_csv('census.csv')
df.head()
```

```python
# In this data set there are two summarized levels, one that contains summary data for the county
# and one that contains summary data for each state. This has been denoted by the SUMLEV feature.

df['SUMLEV'].unique()
```

```python
# Accessing only county level data

df = df[df['SUMLEV'] == 50]
df.head()
```

```python
# Further making a subset for few columns

columns_to_keep = ['STNAME','CTYNAME','BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013',
                   'BIRTHS2014','BIRTHS2015','POPESTIMATE2010','POPESTIMATE2011',
                   'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']
df = df[columns_to_keep]
df.head()
```

```python
# As the data has been broken down by county and state, lets use a combination of both as the new index

df = df.set_index(['STNAME', 'CTYNAME'])
df
```
