```python
# Instead of letting Pandas create a new index, we can set a column to be index as well.

df = pd.read_csv("Admission_Predict.csv", index_col=0)
df.head(3)
```

# Updating Column Names

```python
# Extracting column names

df.columns
```

```python
# Renaming Columns to remove whitespace as well as updating abbreviations

new_df = df.rename(columns={'SOP': 'Statement of Purpose',
                            'LOR ': 'Letter of Recommendation',
                            'Chance of Admit ':'Chance of Admit'})

new_df.head(3)
```

```python
# Better Method to clean column names


# Best way to remove whitespace from column names is by using the "strip" function that removes
# all white space in a string. As it is a string function, it needs to be passed as mapper parameter to 
# apply to all columns

new_df = df.rename(mapper=str.strip, axis = "columns")
print("Old names : \n",df.columns)
print("\nNew Names : \n",new_df.columns)
```

```python
# Another way is to extract columns names into a list and updating that list

cols = list(df.columns)
cols = [x.lower().strip() for x in cols]   # striping whitespace and and then making lowercase
cols
```

###  APPLY() to update column names

```python
df=pd.read_csv("presidents.csv")
df.head()
```

```python
# Goal is to extract first and last name from the "President" feature

def splitname(row):
    
    row['First']=row['President'].split(" ")[0]
    row['Last']=row['President'].split(" ")[-1]
    return row


df=df.apply(splitname, axis='columns')
df.head()
```
